```python
num_examles = 1
num_classes = 50

def input_values():
    """
    1と0のランダムな配列を返す
    """
    multiple_values = [map(int,'{0:050b}'.format(i))for i in range(2**20)]
    random.shuffle(multiple_values)
    final_values = []
    for value in multiple_values[:num_examles]:
        temp = []
        for number in value:
            temp.append([number])
        final_values.append(temp)
    return final_values
    
```

```python
def output_values(inputs):
    """
    引数としてinput_values()の結果を受け取ることを前提とする
    例
    input = [[0],[1],[1]]
    return :[[]]
    """
    final_values = []
    for value in inputs:
        output_values = [0 for _ in range(num_classes)]
        count = 0
        for i in value:
            count += i[0]
        if count < num_classes:
            output_values[count] = 1
        final_values.append(output_values)
    return final_values
```

```python
a = input_values()
print(a)
```

```python
b= output_values(a)
print(b)
```
