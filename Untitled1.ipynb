```python
class SimpleLSTM:
    def __init__(self,hidden_layer=256,num_epochs=10,length_of_sequence=25,n_input=1,n_out=1):
        self.hidden_layer = hidden_layer
        self.num_epochs = num_epochs
        self.length_of_sequence = length_of_sequence
        self.n_input = n_input
        self.n_out = n_out
        
    def create_model(self):
        
        self.model = Sequential()
        self.model.add(SimpleRNN(self.hidden_layer,
                    input_shape=(self.length_of_sequence, self.n_input)))
        self.model.add(Dense(self.n_out))
#         self.model.add(LSTM(
#             self.hidden_layer,
#             input_shape=(self.length_of_sequence,self.n_input),
#             dropout=0.5,
#             recurrent_dropout=0.5))
#         self.model.add(Dense(self.n_out))
        self.model.add(Activation("linear"))
        optimizer = Adam(
            lr=0.01,
            beta_1=0.9, 
            beta_2=0.999)
        self.model.compile(
            loss="binary_crossentropy",
            optimizer=optimizer,metrics=["accuracy"])
    def get_data(self):
        wave_data = WaveData(T=1000)
        wave_data.generate()
        # wave_data.plot()
        wave_data.add_noise()
        self.x_train,self.x_validation,self.y_train,self.y_validation = wave_data.get_data(maxlen=self.length_of_sequence)
    def fit(self,batch_size=10,epochs=500):
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1)
        self.model.fit(
            self.x_train,
            self.y_train,
            batch_size=batch_size,
            epochs=epochs,
            validation_data=(self.x_validation,self.y_validation),
            callbacks=[early_stopping]
        )
        
        
    
```

```python
simple_lstm = SimpleLSTM()
simple_lstm.create_model()
simple_lstm.get_data()
simple_lstm.fit()
```
